```python
%matplotlib inline
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
# http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
iris = datasets.load_iris()
iris
# list(iris.keys())
```

```python
# numpy [:, 0:1] -> in jeder Reihe das erste Element auswählen
X = iris["data"][:, 0:1]  # sepal length
X
```

```python
# TESTDATENSATZ
# sepal length = 4.8, 
# sepal width = 2.5,
# peta length = 5.3,
# petal width = 2.4 
# is an iris virginica

# sepal length, sepal width, peta length, petal width
X = iris["data"][:, 0:1]  # sepal length
y = (iris["target"] == 2).astype(np.int)  # array of 1 if Iris-Virginica, else 0
log_reg = LogisticRegression()
log_reg.fit(X, y)

# prüfe mit sepal length ob Iris-Virginica 
X_new_test_data = np.array([4.8]).reshape(1, -1)
y_proba_test_data = log_reg.predict_proba(X_new_test_data)

# random Testdatensatz
X_new = np.linspace(0, 10, 1000).reshape(-1, 1)

y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]
# plt.figure(figsize=(8, 3))
plt.plot([decision_boundary, decision_boundary], [0, 1], "k:", linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.plot(X_new_test_data, y_proba_test_data[:, 1], 'r+')
plt.plot(X_new_test_data, y_proba_test_data[:, 0], 'r+')

# laut Wahrscheinlichkeit des Kriteriums "sepal length" ist es keine Iris-Virginica 
```
