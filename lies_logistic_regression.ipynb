```python
# TESTDATENSATZ
# sepal length = 4.8, 
# sepal width = 2.5,
# peta length = 5.3,
# petal width = 2.4 
# is an iris virginica

# sepal length, sepal width, peta length, petal width
X = iris["data"][:, 0:1]  # sepal length
y = (iris["target"] == 2).astype(np.int)  # array of 1 if Iris-Virginica, else 0
log_reg = LogisticRegression()
log_reg.fit(X, y)

# prüfe mit sepal length ob Iris-Virginica 
X_new_test_data = np.array([4.8]).reshape(1, -1)
y_proba_test_data = log_reg.predict_proba(X_new_test_data)

# random Testdatensatz
X_new = np.linspace(0, 10, 1000).reshape(-1, 1)

y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]
# plt.figure(figsize=(8, 3))
plt.plot([decision_boundary, decision_boundary], [0, 1], "k:", linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.plot(X_new_test_data, y_proba_test_data[:, 1], 'r+')
plt.plot(X_new_test_data, y_proba_test_data[:, 0], 'r+')

# laut Wahrscheinlichkeit des Kriteriums "sepal length" ist es keine Iris-Virginica 
```

```python
# TESTDATENSATZ
# sepal length = 4.8, 
# sepal width = 2.5,
# peta length = 5.3,
# petal width = 2.4 
# is an iris virginica

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.int)

log_reg = LogisticRegression(C=10**10, random_state=42)
log_reg.fit(X, y)

b0, b1 = np.meshgrid(
        np.array([5.3]).reshape(-1, 1),
        np.array([2.4]).reshape(-1, 1),
    )
X_new_test_data = np.c_[b0.ravel(), b1.ravel()]
y_proba_test_data = log_reg.predict_proba(X_new_test_data)

x0, x1 = np.meshgrid(
        np.linspace(2.9, 7, 500).reshape(-1, 1),
        np.linspace(0.8, 2.7, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]

y_proba = log_reg.predict_proba(X_new)

plt.figure(figsize=(10, 4))

plt.plot(X_new_test_data, y_proba_test_data[:, 1], 'r+')
plt.plot(X_new_test_data, y_proba_test_data[:, 0], 'r+')

plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")

zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)


left_right = np.array([2.9, 7])
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]

plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.text(3.5, 1.5, "Not Iris-Virginica", fontsize=14, color="b", ha="center")
plt.text(6.5, 2.3, "Iris-Virginica", fontsize=14, color="g", ha="center")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.axis([2.9, 7, 0.8, 2.7])
plt.show()
```
